```python
def getScale(t):
    
    print(f"Present value is {t}")
    root.config(bg=f"#{str(t)[:2]}{str(t)[:2]}{str(t)[:2]}")
    

root = Tk()

Scale(root,from_=10, to=500,highlightbackground="red",
     command=getScale).pack()
Scale(root,from_=10, to=500,orient=HORIZONTAL,showvalue=False).pack()




root.mainloop()
```

```python
def shiftLetter(letter):
    alpha=list("abcdefghijklmnopqrstuvwxyz")
    alpha.extend([" ",","])
    index=(alpha.index(letter.lower())+2)%28
    return alpha[index]
```

```python
shiftLetter(",")
```

```python
word = input("Enter a word: ")
word_trans=""
for letter in word:
    word_trans+= shiftLetter(letter)
print(word_trans)
```

```python
def encrypt(sentence):
    sentence_trans=""
    for letter in sentence:
        sentence_trans+= shiftLetter(letter)
    return sentence_trans
```

```python
encrypt("Dan and Ruth did not show up today")
```

```python
def shiftLetter(letter,by=2):
    alpha=list("abcdefghijklmnopqrstuvwxyz")
    alpha.extend([" ",","])
    index=(alpha.index(letter.lower())+by)%len(alpha)
    return alpha[index]


def decrypt(sentence):
    sentence_trans=""
    for letter in sentence:
        sentence_trans+= shiftLetter(letter,-2)
    return sentence_trans
```

```python
decrypt("fcpacpfatwvjafkfapqvaujqyawravqfc ")
```
